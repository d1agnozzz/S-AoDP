–ö–∞–∂–¥—ã–π –Ω–∞–±–æ—Ä –æ–ø–∏—Å—ã–≤–∞–µ—Ç—Å—è –µ–¥–∏–Ω—Å—Ç–≤–µ–Ω–Ω–æ–π —Å—Ç—Ä–æ–∫–æ–π, –≤ –∫–æ—Ç–æ—Ä–æ–π —Å–æ–¥–µ—Ä–∂–∏—Ç—Å—è —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ ùëõ (2‚â§ùëõ‚â§1000).

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
–í—ã–≤–µ–¥–∏—Ç–µ ùë° —Å—Ç—Ä–æ–∫. –û—á–µ—Ä–µ–¥–Ω–∞—è —Å—Ç—Ä–æ–∫–∞ –¥–æ–ª–∂–Ω–∞ —Å–æ–¥–µ—Ä–∂–∞—Ç—å -1, –µ—Å–ª–∏ –¥–ª—è —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤—É—é—â–µ–≥–æ –Ω–∞–±–æ—Ä–∞ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏—Å–∫–æ–º–æ–π –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –Ω–µ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç. –í –ø—Ä–æ—Ç–∏–≤–Ω–æ–º —Å–ª—É—á–∞–µ –≤—ã–≤–µ–¥–∏—Ç–µ –∏—Å–∫–æ–º—É—é –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫—É. –ï—Å–ª–∏ —Ç–∞–∫–∏—Ö –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫ —Å—É—â–µ—Å—Ç–≤—É–µ—Ç –Ω–µ—Å–∫–æ–ª—å–∫–æ, —Ç–æ –≤—ã–≤–µ–¥–∏—Ç–µ –ª—é–±—É—é –∏–∑ –Ω–∏—Ö.

–ü—Ä–∏–º–µ—Ä\

5 1 3 6 2 4 7\
13 9 7 11 8 4 1 3 5 2 6 10 12 


```python
def permutations(num: int):
    if num < 4:
        return '-1'
    if num == 4:
        return '3 1 4 2'
    else: 
        # odds in descending order
        odd = list()
        for i in range(1, num+1, 2):
            odd.append(i)
        odd.reverse()
        
        # evens in ascending order, but 4 and 2 swwaped
        even = [4, 2]
        for i in range(6, num+1, 2):
            even.append(i)


        return " ".join(map(str, odd+even))

print(permutations(10))
print(permutations(2))
print(permutations(4))
print(permutations(6))
print(permutations(7))
print(permutations(13))

quant = int(input())
inputs = list()
for i in range(quant):
    inputs.append(int(input()))
for i in inputs:
    print(permutations(i))
```

–ì–∞—Ä–∞–Ω—Ç–∏—Ä—É–µ—Ç—Å—è, —á—Ç–æ —Å—É–º–º–∞ ùëõ –ø–æ –≤—Å–µ–º –Ω–∞–±–æ—Ä–∞–º —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 2‚ãÖ105 (‚àëùëõ‚â§2‚ãÖ105).

```python
import heapq


def fill_zero_array(num) :
    array = [0] * num
    heap = [(-num, 0)] # (length, leftInd)

    for i in range(1, num+1):
        length, l = heapq.heappop(heap)
        length = - length
        r = l + length - 1
        mid = (l+r) //2
        array[mid] = i

        if mid > l:
            heapq.heappush(heap, (-(mid-l), l))
        if mid < r:
            heapq.heappush(heap, (-(r-mid), mid+1))
    return array


print(*fill_zero_array(1))
print(*fill_zero_array(2))
print(*fill_zero_array(3))
print(*fill_zero_array(4))
print(*fill_zero_array(5))
print(*fill_zero_array(6))

quant = int(input())
inputs = list()
for i in range(quant):
    inputs.append(int(input()))
for i in inputs:
    print(*fill_zero_array(i))
```

–í—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
–í –ø–µ—Ä–≤–æ–π —Å—Ç—Ä–æ–∫–µ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö —Å–æ–¥–µ—Ä–∂–∏—Ç—Å—è –Ω–µ–ø—É—Å—Ç–∞—è —Å—Ç—Ä–æ–∫–∞, –æ–ø–∏—Å—ã–≤–∞—é—â–∞—è —Ä–µ—Ü–µ–ø—Ç ¬´–≥–∞–º–±—É—Ä–≥–µ—Ä–∞ –æ—Ç –ü–æ–ª–∏–∫–∞—Ä–ø–∞¬ª. –î–ª–∏–Ω–∞ —Å—Ç—Ä–æ–∫–∏ –Ω–µ –ø—Ä–µ–≤–æ—Å—Ö–æ–¥–∏—Ç 100, —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Ç–æ–ª—å–∫–æ –±—É–∫–≤—ã 'B' (–ø—Ä–æ–ø–∏—Å–Ω–∞—è –ª–∞—Ç–∏–Ω—Å–∫–∞—è B), 'S' (–ø—Ä–æ–ø–∏—Å–Ω–∞—è –ª–∞—Ç–∏–Ω—Å–∫–∞—è S) –∏ 'C' (–ø—Ä–æ–ø–∏—Å–Ω–∞—è –ª–∞—Ç–∏–Ω—Å–∫–∞—è C).

–í—Ç–æ—Ä–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Ç—Ä–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ nb, ns, nc (1‚Äâ‚â§‚Äânb,‚Äâns,‚Äânc‚Äâ‚â§‚Äâ100) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∫—É—Å–æ—á–∫–æ–≤ —Ö–ª–µ–±–∞, –∫–æ–ª–±–∞—Å—ã –∏ —Å—ã—Ä–∞ –Ω–∞ –∫—É—Ö–Ω–µ –ü–æ–ª–∏–∫–∞—Ä–ø–∞. –¢—Ä–µ—Ç—å—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Ç—Ä–∏ —Ü–µ–ª—ã—Ö —á–∏—Å–ª–∞ pb, ps, pc (1‚Äâ‚â§‚Äâpb,‚Äâps,‚Äâpc‚Äâ‚â§‚Äâ100) ‚Äî —Ü–µ–Ω–∞ –æ–¥–Ω–æ–≥–æ –∫—É—Å–æ—á–∫–∞ —Ö–ª–µ–±–∞, –∫–æ–ª–±–∞—Å—ã –∏ —Å—ã—Ä–∞ –≤ –º–∞–≥–∞–∑–∏–Ω–µ. –ù–∞–∫–æ–Ω–µ—Ü, —á–µ—Ç–≤–µ—Ä—Ç–∞—è —Å—Ç—Ä–æ–∫–∞ —Å–æ–¥–µ—Ä–∂–∏—Ç —Ü–µ–ª–æ–µ —á–∏—Å–ª–æ r (1‚Äâ‚â§‚Äâr‚Äâ‚â§‚Äâ1012) ‚Äî –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä—É–±–ª–µ–π —É –ü–æ–ª–∏–∫–∞—Ä–ø–∞.

–ü–æ–∂–∞–ª—É–π—Å—Ç–∞, –Ω–µ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ —Å–ø–µ—Ü–∏—Ñ–∏–∫–∞—Ç–æ—Ä %lld –¥–ª—è —á—Ç–µ–Ω–∏—è –∏–ª–∏ –∑–∞–ø–∏—Å–∏ 64-–±–∏—Ç–Ω—ã—Ö —á–∏—Å–µ–ª –Ω–∞ –°++. –†–µ–∫–æ–º–µ–Ω–¥—É–µ—Ç—Å—è –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å –ø–æ—Ç–æ–∫–∏ cin, cout –∏–ª–∏ —Å–ø–µ—Ü–∏—Ñ–∏–∫–∞—Ç–æ—Ä %I64d.

–í—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
–í—ã–≤–µ–¥–∏—Ç–µ –Ω–∞–∏–±–æ–ª—å—à–µ–µ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≥–∞–º–±—É—Ä–≥–µ—Ä–æ–≤, –∫–æ—Ç–æ—Ä–æ–µ –ü–æ–ª–∏–∫–∞—Ä–ø –º–æ–∂–µ—Ç –ø—Ä–∏–≥–æ—Ç–æ–≤–∏—Ç—å. –ï—Å–ª–∏ –æ–Ω –Ω–µ –º–æ–∂–µ—Ç –ø—Ä–∏–≥–æ—Ç–æ–≤–∏—Ç—å –Ω–∏ –æ–¥–Ω–æ–≥–æ –≥–∞–º–±—É—Ä–≥–µ—Ä–∞, –≤—ã–≤–µ–¥–∏—Ç–µ 0.

–ü—Ä–∏–º–µ—Ä—ã\

1 2 3\
4\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\

–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
BBC\

1 10 1\
21\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
7\
–≤—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
BSC\
1 1 1\
1 1 3\
1000000000000\
–≤—ã—Ö–æ–¥–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ\
200000000001

```python
def is_possible_to_cook(needed, availale, prices, money, guess):
    bread = max(0,guess*needed[0]-availale[0]) * prices[0]
    sausage = max(0,guess*needed[1]-availale[1]) * prices[1]
    cheese = max(0,guess*needed[2]-availale[2]) * prices[2]
 
    return bread+sausage+cheese <= money
    
 
def max_burgers(recipe: str, available, prices, money):
    b_needed = recipe.count('B')
    s_needed = recipe.count('S')
    c_needed = recipe.count('C')
 
    needed = (b_needed, s_needed, c_needed)
 
    guess = sum(available) + money
 
    l = 0
    r = guess
 
    while r >= l:
        mid = (l + r) // 2
 
        if is_possible_to_cook(needed, available, prices, money, mid):
            l = mid + 1
        else:
            r = mid - 1
    return r

print(max_burgers('BBBSSC', [6, 4, 1], [1, 2, 3], 4))
print(max_burgers('BBC', [1, 10, 1], [1, 10, 1], 21))
print(max_burgers('BSC', [1, 1, 1], [1, 1, 3], 1000000000000))

if __name__ == "__main__":
    recipe = input()
    available = list(map(int, input().split(" ")))
    prices = list(map(int, input().split(" ")))
    money = int(input())

    print(max_burgers(recipe, available, prices, money))
```

–í –ø–µ—Ä–≤–æ–º –Ω–∞–±–æ—Ä–µ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö Ayush –º–æ–∂–µ—Ç —É–¥–∞–ª–∏—Ç—å —Ç–æ–ª—å–∫–æ –≤–µ—Ä—à–∏–Ω—É 2 –∏–ª–∏ 3, –ø–æ—Å–ª–µ —á–µ–≥–æ –≤–µ—Ä—à–∏–Ω–∞ 1 —Å—Ç–∞–Ω–æ–≤–∏—Ç—Å—è –ª–∏—Å—Ç–æ–º, –∏ Ashish –º–æ–∂–µ—Ç —É–¥–∞–ª–∏—Ç—å –µ–µ –≤ —Å–≤–æ—é –æ—á–µ—Ä–µ–¥—å.

–í–æ –≤—Ç–æ—Ä–æ–º –Ω–∞–±–æ—Ä–µ –≤—Ö–æ–¥–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö Ayush –º–æ–∂–µ—Ç —É–¥–∞–ª–∏—Ç—å –≤–µ—Ä—à–∏–Ω—É 2 –Ω–∞ —Å–∞–º–æ–º –ø–µ—Ä–≤–æ–º —à–∞–≥–µ.

```python
from graph2 import Graph

def tree_game(num_of_vertices: int, goal: int, vertices):
    # games = int(input())
    # for _ in range(games):
    #     num_of_vertices, goal = list(map(int, input().split(" ")))
    tree = Graph()
    for vert in vertices:
        u, v = tuple(map(str, vert))
        tree.add_edge_not_oriented(u, v, 1)

    if num_of_vertices == 1 or len(tree.neighbors(str(goal))) == 1:
        print("Ayush")
        return

    if num_of_vertices % 2 == 1:
        print("Ashish")
    else:
        print("Ayush")

tree_game(3, 1, [(2,1),(3,1)])
tree_game(3, 2, [(1,2),(1,3)])

games = int(input())
for _ in range(games):
    num_of_vertices, goal = list(map(int, input().split(" ")))
    edges = []
    for _ in range(num_of_vertices-1):
        u, v = input().split(" ")
        edges.append((u, v))
    tree_game(num_of_vertices, goal, edges)
```

```python
from math import inf

def monster_bullets(array):
    bullets = 0
    prev_exp = array[-1][-1]
    candidate = 0

    mini = inf

    for hp, exp in array:
        if hp > prev_exp:
            bullets += hp - prev_exp
            candidate = prev_exp
        else:
            candidate = hp

        if candidate < mini:
            mini = candidate

        prev_exp = exp
    return bullets + mini

print(monster_bullets([(7, 15), (2, 14), (5, 3)]))   

games = int(input())
for _ in range(games):
    monsters = int(input())
    characteristics = list()
    for monster in range(monsters):
        characteristics.append(tuple(map(int, input().split(' '))))
    print(monster_bullets(characteristics)) 
```

```python
def hilberts_hotel(n, array):
    s = set()
    for k in range(n):
        new_room = (k + array[k])%n 
        if new_room in s:
            return "NO"
        s.add(new_room)
    return "YES"


print(hilberts_hotel(1, [14]))
print(hilberts_hotel(2, [-1, 1]))
print(hilberts_hotel(4, [5, 5, 5, 1]))
print(hilberts_hotel(3, [3, 2, 1]))
print(hilberts_hotel(2, [0, 1]))
print(hilberts_hotel(5, [-239, -2, -100, -3, -11]))
# print(hilberts_hotel(21, list(map(int, "-875686917 410529943 -707495183 127424171 247251083 130195224 -838443642 -833434989 -908630073 36578008 -158304369 -401354971 -932857686 34772737 -960777984 -817720620 -955394591 -918340921 276591458 -681756438 944119892".split(' ')))))

data_size = int(input())
for _ in range(data_size):
    n = int(input())
    a = list(map(int, input().split(' ')))
    print(hilberts_hotel(n, a))
```

–í –ø–µ—Ä–≤–æ–º –ø—Ä–∏–º–µ—Ä–µ ùë°={lcm({1,1})}={1}, –∏ gcd(ùë°)=1.

–í–æ –≤—Ç–æ—Ä–æ–º –ø—Ä–∏–º–µ—Ä–µ ùë°={120,40,80,120,240,80}. –ù–µ—Ç—Ä—É–¥–Ω–æ –≤–∏–¥–µ—Ç—å, —á—Ç–æ gcd(ùë°)=40.

```python
def gcd(a: int, b: int):
    if a == 0:
        return b
    return gcd(b%a, a)

def lcm(a: int, b: int):
    return (a*b)//gcd(a, b)

# pair generator
def create_pairs(n, array):
    for i in range(n):
        for j in range(i+1, n):
            yield lcm(array[i], array[j])

def orak_lcm(n: int, array):

    # pair_lcm = list()
    # for i, x in enumerate(array):
    #     for j, y in enumerate(array[i+1:]):
    #         pair_lcm.append(lcm(x, y))

    pair_gen = create_pairs(n, array)
    _gcd = 0
    for i in pair_gen:
        cur_gcd = gcd(i, next(pair_gen, 0))
        _gcd = gcd(_gcd, cur_gcd)
    return _gcd

print(orak_lcm(4, [10, 24, 40, 80]))
print(orak_lcm(10, [540, 648, 810, 648, 720, 540, 594, 864, 972, 648]))

n = int(input())
array = list(map(int, input().split()))
print(orak_lcm(n, array))
```

```python
def reverso_lexycograph(n, costs, strs):
    r_strs = [strs[i][::-1] for i in range(n)]  # reversed

    from math import inf
    maxx = inf

    dp = [[maxx, maxx] for i in range(n)]
    # not reversed
    dp[0][0] = 0
    # reversed (cost wasted)
    dp[0][1] = costs[0]

    for i in range(1, n):

        if r_strs[i] >= strs[i - 1]:
            dp[i][1] = dp[i - 1][0] + costs[i]
        if r_strs[i] >= r_strs[i - 1]:
            dp[i][1] = min(dp[i][1], dp[i - 1][1] + costs[i])
        if strs[i] >= r_strs[i - 1]:
            dp[i][0] = dp[i - 1][1]
        if strs[i] >= strs[i - 1]:
            dp[i][0] = min(dp[i][0], dp[i - 1][0])
    ans = min(dp[-1])
    if ans >= maxx:
        return -1
    else:
        return ans

print(reverso_lexycograph(2, [1, 2], ['ba', 'ac']))
print(reverso_lexycograph(3, [1, 3, 1], ['aa', 'ba', 'ac']))
print(reverso_lexycograph(2, [5, 5], ['bbb', 'aaa']))
print(reverso_lexycograph(2, [3, 3], ['aaa', 'aa']))


n = int(input())
costs = list(map(int, input().split()))

strs = [input() for i in range(n)]
print(reverso_lexycograph(n, costs, strs))
```

–í —à–µ—Å—Ç–æ–º –Ω–∞–±–æ—Ä–µ, –æ–¥–Ω–æ –∏–∑ –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã—Ö —Ä–µ—à–µ–Ω–∏–π ‚Äî —Ä–∞–∑–±–∏—Ç—å phoenix –Ω–∞ ehinopx.

```python
def phoenix(n, k, string):
    sstr = ''.join(sorted(string))

    if k == 1:
        return ''.join(sstr)

    # if min(sstr) occurencies less than k or k = n 
    if sstr[0] != sstr[k-1] or n == k :
        return sstr[k-1]

    ans = sstr[0]

    # if remaining are NOT the same
    if sstr[k] != sstr[n-1]:   
        ans += sstr[k:n]

    else:
        # if same, distribute evenly
        for _ in range((n-1)//k):
            ans += sstr[n-1]

    return ans

print(phoenix(4, 2, 'baba'))
print(phoenix(5, 2, 'baacb'))
print(phoenix(5, 3, 'baacb'))
print(phoenix(5, 3, 'aaaaa'))
print(phoenix(6, 4, 'aaxxzz'))
print(phoenix(7, 1, 'phoenix'))


data_size = int(input())
for _ in range(data_size):
    n, k = list(map(int, input().split()))
    s = input()
    print(phoenix(n, k, s))
```
